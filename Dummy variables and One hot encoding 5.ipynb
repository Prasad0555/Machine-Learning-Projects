```python
import pandas as pd
df = pd.read_csv('https://github.com/codebasics/py/raw/master/ML/5_one_hot_encoding/homeprices.csv')
df
```

```python
df.info()
```

```python
# creating dummy variables
dummies = pd.get_dummies(df.town)
dummies
```

```python
# concatinate dummies varaible with our original dataframe
merged = pd.concat([df,dummies],axis='columns')  # we merged dummies variable to original dataframe columnwise
merged
```

Dummy Variable Trap

When you can derive one variable from other variables, they are known to be multi-colinear. Here if you know values of 
california and georgia then you can easily infer value of new jersey state, i.e. california=0 and georgia=0. There for these
state variables are called to be multi-colinear. In this situation linear regression won't work as expected. Hence you need 
to drop one column

NOTE: sklearn library takes care of dummy variable trap hence even if you don't drop one of the state columns it is 
    going to work, however we should make a habit of taking care of dummy variable trap ourselves just in case library 
    that you are using is not handling this for you

```python
# dropping town and west windsor column
final = merged.drop(['town','west windsor'],axis='columns')
final
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
x = final.drop('price',axis='columns')   # inputs variables
x
```

```python
y = final.price  # output 
y
```

```python
model.fit(x,y)  # when you do model.fit it means you are going to train the model
```

```python
model.predict([[2800,0,1]]) # here we are going to predict price of 2800sqrt area in robinsville so we put 0 for monroe township
                        # and 1 for robinsville
```

```python
model.predict([[3400,0,0]]) # here we are going to predict price of west windsor so we put 0,0 there
```

```python
model.score(x,y)
```

# Using sklearn OneHotEncoder

First step is to use label encoder to convert town names into numbers

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
dfle = df
dfle.town = le.fit_transform(dfle.town)
dfle
```

```python
x = dfle[['town','area']].values  # here we use values because we want x to be two dimensional array not a dataframe
x
```

```python
y = dfle.price
y
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([('town', OneHotEncoder(), [0])], remainder = 'passthrough')
```

```python
x = ct.fit_transform(x)
x
```

```python
# now to avoid the dummy variable trap we are going to drop one of the column so I will just drop 1st column
# The way you that into dimensional array is -
x = x[:,1:]   # here take all the rows and for the column take all the column from index 1 onwards this will essencially drop
              # you 0 column
x

# so you have 3 columns now
```

```python
model.fit(x,y)
```

```python
model.predict([[0,1,3400]]) # 3400 sqr ft home in west windsor
```

```python
model.predict([[1,0,2800]]) # 2800 sqr ft home in robbinsville
```
